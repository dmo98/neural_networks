```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import cv2
import tqdm
from tqdm import tqdm 
import random

print('TensorFlow version: ', tf.__version__)
print('Keras version: ', keras.__version__)
```

### Cats vs Dogs Image Classifier using a CNN built with Keras's Sequential API

```python
DATA_DIR = 'DATA/PetImages'
CATEGORIES = ['Dog', 'Cat']
IMG_SIZE = 64
CHANNELS = 3
```

```python
def create_training_data():
    training_data = []
    for category in CATEGORIES:
        path = os.path.join(DATA_DIR, category) # path to cats or dogs dir
        if category == 'Dog':
            label = np.array([1,0])
        elif category == 'Cat':
            label = np.array([0,1])
        #label = CATEGORIES.index(category)
        for img_file in tqdm(os.listdir(path)):
            try:
                img_array = cv2.resize(cv2.imread(os.path.join(path, img_file)), (IMG_SIZE,IMG_SIZE))
                training_data.append([img_array, label])
            except Exception as e: # for os errors
                pass
            
    return training_data
```

```python

try:
    X = np.load('features_rgb.npy')
    Y = np.load('labels_2.npy')
except:

    training_data = create_training_data()
    random.shuffle(training_data)

    X = []
    Y = []
    for features, labels in training_data:
        X.append(features)
        Y.append(labels)
    X = np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,CHANNELS)
    Y = np.array(Y)

    # Save the features and labels 
    np.save('features_rgb.npy', X)
    np.save('labels_2.npy', Y)
```

```python
train_set_size = int(0.8*X.shape[0])
valid_set_size = int(0.1*X.shape[0])
test_set_size = X.shape[0] - train_set_size - valid_set_size

X_train = X[:train_set_size]/255.0
Y_train = Y[:train_set_size]
X_valid = X[train_set_size:train_set_size+valid_set_size]/255.0
Y_valid = Y[train_set_size:train_set_size+valid_set_size]
X_test = X[train_set_size+valid_set_size:]/255.0
Y_test = Y[train_set_size+valid_set_size:
          ]
print('Train Set: ', X_train.shape, Y_train.shape)
print('Valid Set: ', X_valid.shape, Y_valid.shape)
print('Test Set: ', X_test.shape, Y_test.shape)
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(64, (3,3), activation='relu', padding='same', input_shape=[IMG_SIZE,IMG_SIZE,CHANNELS]),
    keras.layers.MaxPooling2D(2),
    keras.layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    keras.layers.MaxPooling2D(2),
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.4),
    keras.layers.Dense(2, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
try:
    model = keras.models.load_model('cats_vs_dogs_CNN-RGB.h5')
except:
    lr_scheduler = keras.callbacks.ReduceLROnPlateau(factor=0.3,patience=2)
    early_stopping_cb = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
    history = model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_valid,Y_valid), verbose=2, callbacks=[early_stopping_cb,lr_scheduler])
    
    pd.DataFrame(history.history).plot(figsize=(8,5))
    plt.grid(True)
    plt.gca().set_ylim(0,1)
    plt.title('Learning Curves')
    plt.xlabel('epochs')
    plt.show()
    
    model.save('cats_vs_dogs_CNN-RGB.h5') # [0.4320684906381827, 0.8108974]
    #model.save('cats_vs_dogs_CNN.h5') # [0.45819003803607744, 0.78405446]
```

```python
model.evaluate(X_test, Y_test)
```

```python
Y_test_proba = model.predict(X_test).round(2) # use predict() to get the estimated probabilities for each class

def output_predictions(Y_proba):
    Y_pred_classes = np.argmax(Y_proba, axis=-1)
    Y_pred_one_hot = keras.utils.to_categorical(Y_pred_classes, num_classes=2)
    
    return Y_pred_classes, Y_pred_one_hot.astype('int32')

Y_test_pred_classes, Y_test_pred = output_predictions(Y_test_proba)

for i in range(20):
    print('{}, {}, \t, {}'.format(Y_test_proba[i], Y_test_pred[i], Y_test[i]))
```

```python
# Plot the images, predicted labels and the target labels
n_rows = 20
n_cols = 5
plt.figure(figsize=(n_cols * 3, n_rows * 3))
for row in range(n_rows):
    for col in range(n_cols):
        index = n_cols * row + col
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(X_test[index][:,:,::-1], interpolation="nearest")
        plt.axis('off')
        if np.argmax(Y_test[index], axis=-1) == Y_test_pred_classes[index]:
            color = 'blue'
        else:
            color = 'red'
        plt.title("{} {:2.0f}% ({})".format(CATEGORIES[Y_test_pred_classes[index]],
                                    100*np.max(Y_test_proba[index]),
                                    CATEGORIES[np.argmax(Y_test[index], axis=-1)]),
                                    color=color)
plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.savefig('cats_vs_dogs_predictions_plot', tight_layout=False)
plt.show()
```
