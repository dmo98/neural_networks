```python
'''This notebook implements a Denoising Convolutional Autoencoder, trained on MNIST data to recover the original images from the
noisy input images. It is implemented using tf.keras.'''
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

print('TensorFlow version: ', tf.__version__)
print('Keras version: ', keras.__version__)
```

```python
def preprocess(X, add_noise=False):
    '''This function performs the following pre-processing steps:
    1. Divide all pixel values by 255 so that the range of values is [0,1]
    2. Reshape each 28*28 image to a 28*28*1 image, as needed by the encoder of the DAE.
    3. Add noise to the input images of the DAE (add_noise is set to True)
    '''
    X = X.astype('float32')/255.
    X = np.reshape(X, (X.shape[0], 28, 28, 1))
    if add_noise:
        noise_coeff = 1.5e-3
        X = X + noise_coeff * np.random.normal(loc=0.0, scale=1.0, size=X.shape)
        X = np.clip(X, 0., 1.)
    return X

def display(array1, array2, num_images=10, save_figure=False, save_as=''):
    '''This function selects 10 indices randomly (set by the num_images input parameter), and plots the images at these indices from 2 arrays.'''
    indices = np.random.randint(len(array1), size=num_images)
    images1 = array1[indices, :]
    images2 = array2[indices, :]

    plt.figure(figsize=(2*num_images, 4))
    for i, (image1, image2) in enumerate(zip(images1, images2)):
        ax = plt.subplot(2, num_images, i + 1)
        plt.imshow(image1.reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        ax = plt.subplot(2, num_images, i + 1 + num_images)
        plt.imshow(image2.reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    if save_figure:
        plt.savefig(save_as, tight_layout=False)
    plt.show()
    
```

```python
(X_train, _), (X_test, _) = keras.datasets.mnist.load_data() # Load the data. The labels are ignored.

VALID_SET_SIZE = 5000
X_valid, X_test = X_test[:VALID_SET_SIZE], X_test[VALID_SET_SIZE:] # Both the validation and test sets have 5000 images now
X_train, X_valid, X_test = preprocess(X_train), preprocess(X_valid), preprocess(X_test) # Pre-process the data without adding noise. These will be the desired outputs of the DAE.

X_train_noisy, X_valid_noisy, X_test_noisy = preprocess(X_train, add_noise=True), preprocess(X_valid, add_noise=True), preprocess(X_test, add_noise=True) # Pre-process the data and add noise. These will be the noisy inputs of the DAE.

print('Training Set: {}, {}'.format(X_train.shape, X_train_noisy.shape))
print('Validation Set: {}, {}'.format(X_valid.shape, X_valid_noisy.shape))
print('Test Set: {}, {}'.format(X_test.shape, X_test_noisy.shape))

display(X_train, X_train_noisy, num_images=10) # Display 10 images without the noise and with the noise added.
```

```python
try:
    dae = keras.models.load_model('dae_mnist.h5') # Load the model that has been trained already to save time.
except: 
    # The Encoder
    encoder = keras.models.Sequential([
        keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=[28,28,1]),
        keras.layers.MaxPool2D(pool_size=2, padding='same'),
        keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same'),
        keras.layers.MaxPool2D(pool_size=2, padding='same')
    ])
    # The Decoder
    decoder = keras.models.Sequential([
        keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, activation='relu', padding='same', input_shape=[7,7,32]),
        keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, activation='relu', padding='same'),
        keras.layers.Conv2D(1, kernel_size=3, activation='sigmoid', padding='same')
    ])
    # The Denoising Convolutional Autoencoder 
    dae = keras.models.Sequential([encoder, decoder])
    dae.compile(loss='binary_crossentropy', optimizer='adam')
    
    early_stopping_callback = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True) # for early stopping
    history = dae.fit(X_train_noisy, X_train, epochs=20, batch_size=128, shuffle=True, validation_data=(X_valid_noisy, X_valid), callbacks=[early_stopping_callback])
    model.save('dae_mnist.hf')
```
